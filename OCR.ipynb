```python
#importing the important libraries: OpenCV, Tesserect, os, PIL, bpy
import cv2, os, os.path, matplotlib.pyplot as plt
from PIL import Image

imgs = []
path = "C:/Users/Dorian/Desktop/pictures_OCR"
formats = [".jpg",".gif",".png",".tga"]
for i in os.listdir(path):
    extension = os.path.splitext(i)[1]
    if extension.lower() not in formats:
        continue
    imgs.append(Image.open(os.path.join(path,i)))
```
